#### estudiante 1

La distribución acumulada $G(x)$ se encuentra integrando la función de densidad de probabilidad $g(x)$ desde $0$ hasta $x$:

$$G(x) = \int_{0}^{x} g(t) dt = \int_{0}^{x} (300t^5 - 840t^6 + 720t^7 - 180t^8) dt $$

Calculando la integral:

$$G(x) = 60x^6 - 140x^7 + 120x^8 - 45x^9 $$

La función inversa $G^{-1}(u)$ se encuentra despejando $x$ de la ecuación anterior:

$$u = G(x) = 60x^6 - 140x^7 + 120x^8 - 45x^9 $$

Resolviendo para $x$, obtenemos:

$$x = G^{-1}(u) = \frac{1}{9} \ln \left(\frac{20u}{3} + \frac{\sqrt{3}\sqrt{135u^2 - 140u + 16}}{3}\right)$$

El método de la transformada inversa es una técnica que nos permite generar variables aleatorias que siguen una distribución de probabilidad específica a partir de una distribución uniforme en el intervalo [0,1].

Primero, necesitamos calcular la distribución acumulada de la función de densidad g(x). La distribución acumulada se define como:

$$G(x) = \int_{-\infty}^{x} g(t) dt$$

Para nuestra función de densidad g(x) = a sin(πx), la distribución acumulada se puede calcular de la siguiente manera:

$$G(x) = \int_{-\infty}^{x} g(t) dt = \int_{-\infty}^{x} a \sin(\pi t) dt = \frac{-a}{\pi} \cos(\pi x) + \frac{a}{\pi}$$

Ahora, necesitamos invertir la distribución acumulada G(x) para obtener la función inversa G⁻¹(u) que se usa para generar variables aleatorias. La función inversa se puede calcular de la siguiente manera:

$$u = G(x) = \frac{-a}{\pi} \cos(\pi x) + \frac{a}{\pi}$$

Despejando x, obtenemos la función inversa:

$$x = G^{-1}(u) = \frac{1}{\pi} \cos^{-1}(\frac{\pi u - 1}{a})$$

Ahora, podemos generar variables aleatorias que sigan la distribución g(x) usando el siguiente algoritmo:

Generar una variable aleatoria uniforme U en el intervalo [0,1].
Calcular la variable aleatoria que sigue la distribución g(x) como X = G⁻¹(U).

#### estudiante 2
Para obtener la función de distribución acumulada $G(x)$, se integra la función de densidad de probabilidad $g(x)$ desde $0$ hasta $x$. En nuestro caso, $g(x) = 300x^5 - 840x^6 + 720x^7 - 180x^8$, por lo que al calcular la integral se obtiene $G(x) = 60x^6 - 140x^7 + 120x^8 - 45x^9$. Para encontrar la función inversa $G^{-1}(u)$, se despeja $x$ de la ecuación anterior. Al hacerlo, se llega a la expresión 

$x = G^{-1}(u) = \frac{1}{9} \ln \left(\frac{20u}{3} + \frac{\sqrt{3}\sqrt{135u^2 - 140u + 16}}{3}\right)$. En resumen, la función inversa $G^{-1}(u)$ es la solución de la ecuación $u = G(x)$, donde $G(x)$ es la función de distribución acumulada obtenida al integrar $g(x)$ y $u$ es un número aleatorio uniformemente distribuido en el intervalo $[0,1]$.

El método de la transformada inversa es una técnica utilizada para generar variables aleatorias con una distribución de probabilidad específica a partir de una distribución uniforme en el intervalo [0,1]. Para ello, primero es necesario calcular la distribución acumulada de la función de densidad g(x). En este caso, la función de densidad es g(x) = a sin(πx), y su distribución acumulada se obtiene integrando g(x) desde -∞ hasta x.

Luego, se debe invertir la distribución acumulada G(x) para obtener la función inversa G⁻¹(u), que se utiliza para generar variables aleatorias. La función inversa se obtiene despejando x de la ecuación u = G(x), donde u es una variable aleatoria uniforme en el intervalo [0,1].

Finalmente, para generar variables aleatorias que sigan la distribución g(x), se utiliza un algoritmo que consiste en generar una variable aleatoria uniforme U en el intervalo [0,1], y luego calcular la variable aleatoria X = G⁻¹(U), donde G⁻¹ es la función inversa de la distribución acumulada de la función de densidad g(x).

```python
# estudiante 2

def Acep_rechazo(R2, R1, f, t):
    max_f = f(optimize.fmin(lambda x:-f(x), 0, disp=False))
    x = np.arange(0, 1, 0.01)
    plt.plot(x, f(x), label='f(x)')
    plt.plot(x, max_f*t(x), label='t(x)')
    accepted_points = []
    rejected_points = []
    for i in range(len(R1)):
        if R2[i] * max_f * t(R1[i]) <= f(R1[i]):
            accepted_points.append((R1[i], R2[i] * max_f * t(R1[i])))
        else:
            rejected_points.append((R1[i], R2[i] * max_f * t(R1[i])))
    plt.scatter(*zip(*accepted_points), marker='o', color='b', s=1)
    plt.scatter(*zip(*rejected_points), marker='o', color='r', s=1)
    plt.legend()
    plt.show()

# Función objetivo 
f = lambda x: 30 * (x**2 - 2*x**3 + x**4)

# Función t -> Pulso senoidal
t = lambda x: np.sin(np.pi * x)

# Validación del método
N = 10000 # número de puntos a simular
R2 = np.random.rand(N)
R1 = np.random.rand(N)

# Prueba de función creada
Acep_rechazo(R2, R1, f, t)
```

```python
# estudiante 2

def aceptacion_rechazo(N, f, t, x_min, x_max):
    # Encuentra el máximo de f en el rango dado
    max_f = f(optimize.fmin(lambda x: -f(x), (x_max - x_min) / 2, disp=False))

    # Genera puntos aleatorios en el rango dado
    X = np.random.uniform(x_min, x_max, N)

    # Genera puntos aleatorios en el rectángulo t(x) x max_f
    U = np.random.uniform(0, max_f, N)

    # Evalúa f(x) y t(x) en los puntos aleatorios
    fx = f(X)
    tx = t(X)

    # Evalúa la condición de aceptación y rechazo
    accept = U < fx / tx

    # Grafica los puntos aceptados y rechazados
    plt.plot(X[accept], U[accept], 'o', alpha=0.5, label='Aceptados')
    plt.plot(X[~accept], U[~accept], 'o', alpha=0.5, label='Rechazados')
    plt.plot(np.arange(x_min, x_max, 0.01), t(np.arange(x_min, x_max, 0.01)), 'k-', label='t(x)')
    plt.xlabel('x')
    plt.ylabel('f(x) / t(x)')
    plt.legend()
    plt.show()

# Ejemplo 1: t(x) constante
f = lambda x: 30 * (x**2 - 2*x**3 + x**4)
t = lambda x: np.ones_like(x) * max_f
x_min, x_max = 0, 1
N = 10000
aceptacion_rechazo(N, f, t, x_min, x_max)

# Ejemplo 2: t(x) un pulso senoidal
f = lambda x: 30 * (x**2 - 2*x**3 + x**4)
t = lambda x: max_f * np.sin(np.pi * x)
x_min, x_max = 0, 1
N = 10000
aceptacion_rechazo(N, f, t, x_min, x_max)
```
